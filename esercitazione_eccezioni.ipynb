1. **Definizione delle Eccezioni:** Crea due classi di eccezione personalizzate che ereditano dalla classe base `Exception`:
   - `ImportoNonValidoError`: da sollevare se l'importo inserito per un deposito o prelievo è $\le 0$. Deve memorizzare l'importo non valido.
   - `SaldoInsufficienteError`: da sollevare se si tenta di prelevare una cifra superiore al saldo disponibile. Deve memorizzare il saldo attuale e l'importo richiesto.
   - Entrambe le classi devono ridefinire il metodo `__str__` per restituire un messaggio d'errore chiaro e dettagliato.


```python
class ImportoNonValidoError(Exception):
    def __init__(self, importo):
        # Memorizziamo l'importo che ha causato l'errore
        self.importo = importo

    def __str__(self):
        # Messaggio mostrato quando l'eccezione viene stampata
        return f"Errore: importo non valido ({self.importo}). L'importo deve essere maggiore di zero."

class SaldoInsufficienteError(Exception):
    def __init__(self, saldo_attuale, importo_richiesto):
        # Memorizziamo saldo disponibile e importo che si voleva prelevare
        self.saldo_attuale = saldo_attuale
        self.importo_richiesto = importo_richiesto

    def __str__(self):
        # Messaggio mostrato quando l'eccezione viene stampata
        return (
            f"Errore: saldo insufficiente. "
            f"Saldo attuale: {self.saldo_attuale} euro, "
            f"importo richiesto: {self.importo_richiesto} euro."
        )
```

2. **Classe `BankAccount`:** Crea la classe principale con le seguenti caratteristiche:
   - Un attributo per il `saldo`.
   - Un metodo costruttore `__init__` che accetti un parametro facoltativo `saldo_iniziale` (valore di default: `0`).
   - Un metodo `deposito(importo)` che:
     - Sollevi `ImportoNonValidoError` se l'importo è $\le 0$.
     - Aggiorni il saldo e stampi un messaggio di conferma in caso contrario.
   - Un metodo `prelievo(importo)` che:
     - Sollevi `ImportoNonValidoError` se l'importo è $\le 0$.
     - Sollevi `SaldoInsufficienteError` se l'importo è maggiore del saldo disponibile.
     - Aggiorni il saldo deducendo l'importo e stampi un messaggio di conferma in caso contrario.

```python
class BankAccount:
    def __init__(self, saldo_iniziale=0):
        # Il conto parte con un saldo iniziale, di default 0
        self.saldo = saldo_iniziale

    def deposito(self, importo):
        # Un deposito deve essere sempre maggiore di zero
        if importo <= 0:
            raise ImportoNonValidoError(importo)

        # Se l'importo è valido, viene aggiunto al saldo
        self.saldo += importo

        print(f"Deposito di {importo} euro effettuato. Saldo attuale: {self.saldo} euro.")

    def prelievo(self, importo):
        # Un prelievo deve essere sempre maggiore di zero
        if importo <= 0:
            raise ImportoNonValidoError(importo)

        # Non si può prelevare più del saldo disponibile
        if importo > self.saldo:
            raise SaldoInsufficienteError(self.saldo, importo)

        # Se i controlli sono superati, il saldo viene diminuito
        self.saldo -= importo

        print(f"Prelievo di {importo} euro effettuato. Saldo attuale: {self.saldo} euro.")
```

3. **Blocco di Test:** Scrivi un codice di test per verificare il corretto funzionamento della classe, catturando le eccezioni all'interno di blocchi **`try...except`** dedicati.

```python
# Creiamo un conto con saldo iniziale di 100 euro
conto = BankAccount(100)

print("SALDO INIZIALE:")
print(f"{conto.saldo} euro")


# TEST 1: deposito valido
try:
    conto.deposito(50)
except ImportoNonValidoError as errore:
    print(errore)


# TEST 2: deposito non valido
try:
    conto.deposito(-20)
except ImportoNonValidoError as errore:
    print(errore)


# TEST 3: prelievo valido
try:
    conto.prelievo(30)
except ImportoNonValidoError as errore:
    print(errore)
except SaldoInsufficienteError as errore:
    print(errore)


# TEST 4: prelievo non valido
try:
    conto.prelievo(0)
except ImportoNonValidoError as errore:
    print(errore)
except SaldoInsufficienteError as errore:
    print(errore)


# TEST 5: prelievo con saldo insufficiente
try:
    conto.prelievo(1000)
except ImportoNonValidoError as errore:
    print(errore)
except SaldoInsufficienteError as errore:
    print(errore)


print("\nSALDO FINALE:")
print(f"{conto.saldo} euro")
```
