# Gestione di un Conto Bancario con Eccezioni Personalizzate

## Obiettivo
Creare una classe `BankAccount` che rappresenta un conto bancario. La classe deve avere un saldo iniziale e metodi per depositare e prelevare denaro. Il programma deve essere in grado di gestire eccezioni personalizzate per far fronte a situazioni di errore comuni come:
- Fondi insufficienti durante un prelievo.
- Importi di deposito o prelievo non validi (minori o uguali a zero).

---

## Traccia

1. **Definizione delle Eccezioni:** Crea due classi di eccezione personalizzate che ereditano dalla classe base `Exception`:
   - `ImportoNonValidoError`: da sollevare se l'importo inserito per un deposito o prelievo è $\le 0$. Deve memorizzare l'importo non valido.
   - `SaldoInsufficienteError`: da sollevare se si tenta di prelevare una cifra superiore al saldo disponibile. Deve memorizzare il saldo attuale e l'importo richiesto.
   - Entrambe le classi devono ridefinire il metodo `__str__` per restituire un messaggio d'errore chiaro e dettagliato.

2. **Classe `BankAccount`:** Crea la classe principale con le seguenti caratteristiche:
   - Un attributo per il `saldo`.
   - Un metodo costruttore `__init__` che accetti un parametro facoltativo `saldo_iniziale` (valore di default: `0`).
   - Un metodo `deposito(importo)` che:
     - Sollevi `ImportoNonValidoError` se l'importo è $\le 0$.
     - Aggiorni il saldo e stampi un messaggio di conferma in caso contrario.
   - Un metodo `prelievo(importo)` che:
     - Sollevi `ImportoNonValidoError` se l'importo è $\le 0$.
     - Sollevi `SaldoInsufficienteError` se l'importo è maggiore del saldo disponibile.
     - Aggiorni il saldo deducendo l'importo e stampi un messaggio di conferma in caso contrario.

3. **Blocco di Test:** Scrivi un codice di test per verificare il corretto funzionamento della classe, catturando le eccezioni all'interno di blocchi **`try...except`** dedicati.
