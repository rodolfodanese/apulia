## ES2 - Negozio

Dobbiamo creare un sistema semplificato per gestire un negozio e le interazioni con i clienti.

Il negozio deve tenere traccia:

- del nome del negozio;
- della categoria del negozio;
- dei prodotti disponibili;
- della quantità disponibile per ogni prodotto;
- del prezzo unitario di ogni prodotto.

I clienti devono poter gestire un carrello, aggiungendo e rimuovendo prodotti.

### Obiettivo

Creare due classi:

- `Negozio`
- `Cliente`

La classe `Negozio` rappresenta il negozio e il suo inventario.

La classe `Cliente` rappresenta un cliente e il suo carrello.

#### Parte 1 — Classe Negozio

Crea una classe chiamata Negozio.

Ogni negozio deve avere:

| Attributo | Significato | Tipo consigliato |
|---|---|---|
| `_nome` | Nome del negozio | `str` |
| `_categoria` | Categoria del negozio | `str` |
| `_inventario` | Prodotti disponibili nel negozio | `dict` |

L’inventario deve essere un dizionario.

Ogni prodotto deve essere salvato con:

- quantità disponibile;
- prezzo unitario.

Esempio di struttura dell’inventario:
```python
inventario = {
    "Pasta": {
        "quantita": 10,
        "prezzo_unitario": 2.5
    },
    "Riso": {
        "quantita": 5,
        "prezzo_unitario": 3.0
    }
}
```

##### Metodi richiesti per `Negozio`

La classe `Negozio` deve avere questi metodi:

| Metodo | Cosa deve fare |
|---|---|
| `aggiungi_prodotto(nome, quantita, prezzo_unitario)` | Aggiunge un prodotto all’inventario |
| `rimuovi_prodotto(nome, quantita)` | Rimuove una certa quantità di un prodotto |
| `ottieni_inventario()` | Restituisce l’inventario del negozio |
| `calcola_totale_vendite()` | Calcola il valore totale dei prodotti presenti nell’inventario |

#### Parte 2 — Classe `Cliente`

Crea una classe chiamata `Cliente`.

Ogni cliente deve avere:

| Attributo | Significato | Tipo consigliato |
|---|---|---|
| `_nome` | Nome del cliente | `str` |
| `_carrello` | Prodotti nel carrello | `dict` |

Il carrello deve essere un dizionario.

Esempio di struttura del carrello:
```python
carrello = {
    "Pasta": 2,
    "Latte": 3
}
```

| `ottieni_carrello()` | Restituisce il carrello del cliente |
