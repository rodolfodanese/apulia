## ES2 — Eccezioni e gestione degli errori

### Obiettivo dell'esercizio

Crea due classi che usano **try-except** per gestire errori:

| Classe | Cosa deve fare |
|---|---|
| `Calcolatrice` | Esegue operazioni base con gestione di divisione per zero e input invalido |
| `ConversoreTemperatura` | Converte temperature con validazione e eccezioni custom |

Argomenti richiamati:
- blocco try-except;
- eccezioni predefinite (ValueError, ZeroDivisionError, TypeError);
- eccezioni personalizzate;
- raise per sollevare errori;
- OOP con error handling.

#### Parte 1 — Classe Calcolatrice

Definisci una classe chiamata `Calcolatrice`.

La classe deve eseguire operazioni aritmetiche di base.

| Metodo | Cosa deve fare |
|---|---|
| `somma(a, b)` | Ritorna a + b |
| `sottrazione(a, b)` | Ritorna a - b |
| `moltiplicazione(a, b)` | Ritorna a * b |
| `divisione(a, b)` | Ritorna a / b, ma gestisce divisione per zero |

Ciascun metodo deve:

- usare un blocco **try-except**;
- catturare `ZeroDivisionError` se b = 0 (in divisione);
- catturare `TypeError` se gli argomenti non sono numeri;
- stampare un messaggio di errore e ritornare `None` se c'è un'eccezione;
- altrimenti ritornare il risultato.

Esempio di errore:
```
divisione(10, 0) → Errore: impossibile dividere per zero!
divisione(10, "a") → Errore: gli argomenti devono essere numeri!
```

#### Parte 2 — Classe ConversoreTemperatura con eccezione custom

Definisci una classe chiamata `ConversoreTemperatura`.

**Prima, crea un'eccezione personalizzata:**
```python
class TemperaturaInvalida(Exception):
    """Eccezione per temperature fuori intervallo valido."""
    pass
```

La classe `ConversoreTemperatura` deve:

| Metodo | Cosa deve fare |
|---|---|
| `celsius_a_fahrenheit(celsius)` | Converte Celsius a Fahrenheit: F = (C * 9/5) + 32 |
| `fahrenheit_a_celsius(fahrenheit)` | Converte Fahrenheit a Celsius: C = (F - 32) * 5/9 |
| `kelvin_a_celsius(kelvin)` | Converte Kelvin a Celsius: C = K - 273.15 |

Ogni metodo deve:

- usare un blocco **try-except**;
- controllare che la temperatura sia **numerica** (catturare `TypeError`);
- controllare che la temperatura sia **valida** (es: Kelvin >= 0, altrimenti sollevare `TemperaturaInvalida`);
- utilizzare **raise** per sollevare l'eccezione personalizzata;
- ritornare il risultato o `None` se c'è un errore.

Esempio di errore:
```
kelvin_a_celsius(-10) → Errore: la temperatura Kelvin non può essere negativa!
celsius_a_fahrenheit("30") → Errore: la temperatura deve essere un numero!
```
