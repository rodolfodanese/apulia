### 1. Definizione della classe `Libro`

Crea una classe chiamata `Libro` che rappresenti un libro presente nel negozio.

La classe deve avere i seguenti attributi:

- `titolo`: il titolo del libro;
- `autore`: l‚Äôautore del libro;
- `anno_pubblicazione`: l‚Äôanno di pubblicazione del libro.

La classe deve contenere:

- un metodo costruttore `__init__(self, titolo, autore, anno_pubblicazione)` che inizializzi gli attributi dell‚Äôoggetto;
- un metodo `__str__(self)` che restituisca una rappresentazione leggibile del libro.

Il metodo `__str__` deve restituire una stringa nel seguente formato:

```python
"Titolo di Autore (Anno)"
```

Esempio:

```python
"1984 di George Orwell (1949)"
```

````python
class Libro:
    def __init__(self, titolo, autore, anno_pubblicazione):
        # Attributi dell'oggetto Libro
        self.titolo = titolo
        self.autore = autore
        self.anno_pubblicazione = anno_pubblicazione

    def __str__(self):
        # Rappresentazione leggibile dell'oggetto quando viene stampato
        return f"{self.titolo} di {self.autore} ({self.anno_pubblicazione})"
````

### 2. Creazione degli oggetti `Libro`

Crea almeno tre oggetti della classe `Libro`.

Ad esempio:

- `libro1`: *Il signore degli anelli*, J.R.R. Tolkien, 1954;
- `libro2`: *Harry Potter e la pietra filosofale*, J.K. Rowling, 1997;
- `libro3`: *1984*, George Orwell, 1949.


````python
libro1 = Libro("Il signore degli anelli", "J.R.R. Tolkien", 1954)
libro2 = Libro("Harry Potter e la pietra filosofale", "J.K. Rowling", 1997)
libro3 = Libro("1984", "George Orwell", 1949)
````

### 3. Gestione dell‚Äôinventario

Crea una lista chiamata `inventario`.

Aggiungi alla lista tutti i libri creati.

Successivamente, stampa tutti i libri presenti nell‚Äôinventario.

Ogni libro deve essere stampato in modo leggibile, sfruttando il metodo `__str__`.

````python
inventario = []

# Aggiungiamo i libri all'inventario
inventario.append(libro1)
inventario.append(libro2)
inventario.append(libro3)

# Stampiamo tutti i libri presenti nell'inventario
print("INVENTARIO INIZIALE:")

for libro in inventario:
    print(libro)
````

### 4. Creazione del dizionario dei libri

Crea un dizionario chiamato `dizionario_libri`.

Il dizionario deve avere:

- come **chiave** il titolo del libro;
- come **valore** l‚Äôoggetto `Libro` corrispondente.

Esempio logico:

```python
{
    "1984": oggetto_libro_1984
}
```

Dopo aver creato il dizionario, stampa tutti i libri contenuti al suo interno.

````python
dizionario_libri = {}

# La chiave √® il titolo, il valore √® l'oggetto Libro
dizionario_libri[libro1.titolo] = libro1
dizionario_libri[libro2.titolo] = libro2
dizionario_libri[libro3.titolo] = libro3

print("\nDIZIONARIO DEI LIBRI:")

for titolo, libro in dizionario_libri.items():
    print(f"{titolo}: {libro}")
````

### 5. Gestione degli autori con un set

Crea un set chiamato `insieme_autori`.

Aggiungi al set gli autori dei libri presenti nel negozio.

Il set deve essere usato perch√© gli autori non devono comparire pi√π di una volta.

Alla fine, stampa tutti gli autori presenti nel negozio.

````python
insieme_autori = set()

# Il set evita automaticamente gli autori duplicati
insieme_autori.add(libro1.autore)
insieme_autori.add(libro2.autore)
insieme_autori.add(libro3.autore)

print("\nAUTORI PRESENTI NEL NEGOZIO:")

for autore in insieme_autori:
    print(autore)
````

### 6. Gestione della coda delle vendite

Crea una lista chiamata `coda_vendite`.

Aggiungi alcuni libri alla coda delle vendite.

Simula la vendita del primo libro presente nella coda:

1. rimuovi il primo libro dalla coda;
2. stampa il libro venduto;
3. rimuovi il libro venduto anche dall‚Äôinventario.

Alla fine, stampa l‚Äôinventario aggiornato.

````python
coda_vendite = []

# Inseriamo due libri nella coda di vendita
coda_vendite.append(libro2)
coda_vendite.append(libro3)

# Vendiamo il primo libro inserito nella coda
libro_venduto = coda_vendite.pop(0)

print("\nLIBRO VENDUTO:")
print(libro_venduto)

# Rimuoviamo il libro venduto anche dall'inventario
inventario.remove(libro_venduto)

print("\nINVENTARIO AGGIORNATO DOPO LA VENDITA:")

for libro in inventario:
    print(libro)
````
