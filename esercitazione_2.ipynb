## ES1 — Manipolazione stringhe e validazione

### Obiettivo dell'esercizio

Scrivi tre funzioni:

| Funzione | Cosa deve fare |
|---|---|
| `estrai_dominio(email)` | Estrae il dominio da un indirizzo email |
| `valida_email(email)` | Valida un indirizzo email (semplice) |
| `conta_parole(testo, parola)` | Conta quante volte una parola appare in un testo, case-insensitive |

Argomenti richiamati:
- metodi di stringa (`split()`, `lower()`, `count()`, `find()`);
- condizioni e operatori logici;
- cicli `for`;
- return.

#### Parte 1 — Funzione estrai_dominio(email)

Scrivi una funzione chiamata estrai_dominio.

La funzione deve:

- accettare un parametro email (stringa);
- estrarre la parte dopo il `@` (il dominio);
- restituire il dominio;
- usare il metodo `split()` della stringa.

Esempio:
```
estrai_dominio("mario@gmail.com") → "gmail.com"
estrai_dominio("alice@outlook.it") → "outlook.it"
```

#### Parte 2 — Funzione valida_email(email)

Scrivi una funzione chiamata valida_email.

La funzione deve:

- accettare un parametro email (stringa);
- controllare che contenga esattamente un carattere `@`;
- controllare che ci sia almeno un carattere prima dell'`@`;
- controllare che ci sia almeno un punto (`.`) dopo l'`@`;
- restituire `True` se valida, `False` altrimenti;
- usare operatori logici come `and`, `or`.

Esempio:
```
valida_email("mario@gmail.com") → True
valida_email("mario@.com") → False
valida_email("mario.gmail.com") → False
valida_email("mario@@gmail.com") → False
```

#### Parte 3 — Funzione conta_parole(testo, parola)

Scrivi una funzione chiamata conta_parole.

La funzione deve:

- accettare due parametri: `testo` (stringa) e `parola` (stringa);
- scorrere il testo e contare le occorrenze della parola;
- ignorare le differenze tra maiuscole e minuscole (case-insensitive);
- restituire il numero di occorrenze;
- usare il metodo `lower()` per normalizzare il testo;
- usare un ciclo `for` per iterare.

Suggerimento: puoi usare `testo.lower().split()` per ottenere una lista di parole.

Esempio:
```
conta_parole("Il gatto è bello, il gatto è veloce", "gatto") → 2
conta_parole("Ciao ciao a tutti", "CIAO") → 2
```

#### Test richiesti

```python
# Test estrai_dominio
email1 = "mario@gmail.com"
dominio = estrai_dominio(email1)
print(f"Il dominio di {email1} è: {dominio}")

print()

# Test valida_email
email_da_testare = ["mario@gmail.com", "mario.gmail.com", "mario@@gmail.com", "@gmail.com"]
for email in email_da_testare:
    risultato = valida_email(email)
    print(f"{email}: {risultato}")

print()

# Test conta_parole
testo = "Python è potente, Python è flessibile, Python è divertente"
parola_cercata = "python"
occorrenze = conta_parole(testo, parola_cercata)
print(f"La parola '{parola_cercata}' appare {occorrenze} volte nel testo.")
```
